```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
n = 20
```

```python
r = nx.random_regular_graph(d = 12, n = n)
```

```python
nx.draw(r)
plt.show()
```

```python
fileName = 'r3.in'
nx.write_adjlist(r, fileName, delimiter = ',', comments = '')
with open(fileName, 'r+') as file:
    lines = file.readlines()[3:]
    lines.insert(0, '{}\n'.format(n))
    file.seek(0)
    file.writelines(lines)
```
